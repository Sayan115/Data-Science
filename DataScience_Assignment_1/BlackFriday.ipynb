### Part 1

```python
import pandas as pd
cf=pd.read_csv("BlackFriday.csv")
cf.head()
```

### Part 2

```python
customer_dataset = cf[['User_ID','Gender','Age']].copy()
customer_dataset=customer_dataset.drop_duplicates()
s=customer_dataset.groupby('Gender').size()
print(s)
```

### Part 3

```python
user_purchase = cf.groupby(['User_ID','Gender'])['Purchase'].sum().reset_index()
user_purchase.groupby('Gender')['Purchase'].mean()
```

### Part 4

```python
max_purchase = cf.groupby(['Age', 'City_Category'])['Purchase'].max().reset_index()
max_purchase.sort_values(by='Purchase', ascending=True)
```

```python
max_purchase = cf.groupby(['Gender', 'Marital_Status'])['Purchase'].max().reset_index()
max_purchase.sort_values(by='Purchase', ascending=False)
```
