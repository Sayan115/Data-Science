```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset("tips")
tips
```

```python
y=tips["tip"]
X1=tips["total_bill"]
```

```python
theta_1=((y.mean()-y)*(X1.mean()-X1)).sum()/((X1.mean()-X1)**2).sum() # coeffcient/Slope
theta_0=y.mean()-X1.mean()*theta_1 # intercept
```

```python
print("Coeff/Slope ",theta_1)
print("Intercept ",theta_0)
y_hat=theta_0+theta_1*X1 # calculating the predicted values using regression equation
```

```python
r=np.corrcoef(X1,y)[0,1] # calculating the correlation coefficient (r) between X1 and y
plt.figure(dpi=150)
sns.scatterplot(data=tips,x=X1,y=y,label=f"{r:.2f}")
plt.plot(X1,y_hat,"r-",label="regression line")
plt.xlabel("Total Bill")
plt.ylabel("Tips")
plt.title("Total Bill vs Tips")
plt.legend()
plt.grid(alpha=0.5, ls='--')
plt.show()
```
